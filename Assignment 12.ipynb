```python
import numpy as np
import pandas as pd
from pandas import DataFrame, Series
import sqlite3 as db
from pandasql import sqldf
pysqldf = lambda q: sqldf(q, globals())
```

```python
adultdata=pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data',names=['Age','Workclass','fnlwgt','Education','Education_Num','Marital_Status','Occupation','Relationship','Race','Sex','Capital_Gain','Capital_Loss','Hours_per_Week','Native_Country','>50K, <=50K'])
```

```python
#1 select 10 records and display
#df.head(10)
pysqldf("SELECT * FROM adultdata LIMIT 10;")
```

```python
def trimAllColumns(df):
    trimStrings = lambda x: x.strip() if type(x) is str else x
    return df.applymap(trimStrings)
adultdata = trimAllColumns(adultdata)
```

```python
#2 show average hours/week of all men in private sector
pysqldf("SELECT  Sex,Workclass,avg(Hours_per_Week) FROM  adultdata WHERE Workclass='Private' and Sex='Male';")
```

```python
#4 married people working in private sector with a masters degree
q ="""SELECT * FROM adultdata
        WHERE Workclass = 'Private' AND 
        Marital_Status LIKE 'Married%' AND
        Education = 'Masters'
            
"""
pysqldf(q)
```

```python
#5 avg, max, min ages for working class groups
q = """SELECT Workclass,
            avg(Age),
            max(Age),
            min(Age)
            FROM adultdata
            GROUP BY Workclass
        
"""
pysqldf(q)
```

```python
#6 age distribution by country
q= """SELECT Native_Country, min(Age),  max(Age) FROM adultdata GROUP BY Native_Country
"""
pysqldf(q)
```
